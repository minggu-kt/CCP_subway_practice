```python
import json
import urllib.request
geo_path = 'https://raw.githubusercontent.com/southkorea/southkorea-maps/master/kostat/2013/json/skorea_submunicipalities_geo_simple.json'
geo_str = json.loads(urllib.request.urlopen(geo_path).read().decode('utf-8'))
```

```python
print(geo_str)
```

```python
len(geo_str['features'])
```

```python
geo_str['features'][0]['properties']['name']
```

```python
geo_str['features'][0]['geometry']['coordinates'][0] # 개수는 다 다름
```

```python
geo_str_s_idx = []
for i in range(3482):
  if geo_str['features'][i]['properties']['name'] in seouldong:
    cnt = 0
    for j in range(len(geo_str['features'][i]['geometry']['coordinates'][0])):
      if geo_str['features'][i]['geometry']['coordinates'][0][j][1] >= 37.5:
        if geo_str['features'][i]['geometry']['coordinates'][0][j][1] <= 37.7:
          if geo_str['features'][i]['geometry']['coordinates'][0][j][0] >= 126.8:
            if geo_str['features'][i]['geometry']['coordinates'][0][j][1] <= 127.0:
              cnt = cnt + 1
    if cnt == len(geo_str['features'][i]['geometry']['coordinates'][0]):
      geo_str_s_idx.append(i)
```

```python
print(len(geo_str_s_idx))
print(geo_str_s_idx)
```

```python
print(geo_str['features'][0])
```

```python
features_list = []
for i in range(162):
  features_list.append(geo_str['features'][geo_str_s_idx[i]])
geo_str_s = {'type': 'FeatureCollection','features': features_list,'type':'FeatureCollection'}
```
