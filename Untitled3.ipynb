```python
import pandas as pd
df1 = pd.read_excel('seoulmetro.xlsx')
df1.info()
```

```python
import seaborn as sns
import matplotlib as mpl
mpl.rc('font', family = 'Malgun Gothic')

sns.countplot(data = df1, y = '노선명')

plt.figure(figsize = [5,10])
```

```python
cond1 = (df1['노선명'] == '1호선')
df2 = df1[cond1]

sns.histplot(data = df2, x = '하차총승객수')
```

```python
sns.scatterplot(data = df2, x = '승차총승객수', y = '하차총승객수')
```
