```python
# 기본적인 library
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rc('font', family = 'Malgun Gothic')
```

```python
df1 = pd.read_csv('datastation3.csv')
df1.head()
```

```python
df1.info()
```

# 다변수

* X : 연숙/ Y : 연속
   - 산점도 / 회귀선/ 상관분석

```python
sns.scatterplot(data = df1, x = 'age', y = 'Mileage')
```

```python
sns.scatterplot(data = df1, x = 'age', y = 'Mileage', hue = 'type_of_contract')
```

```python
sns.scatterplot(data = df1, x = 'age', y = 'Mileage', hue = 'type_of_contract', size = 'Term')
```

```python
# 회귀선 그리기
sns.lmplot(data = df1, x = 'age', y = 'Mileage', hue = 'type_of_contract')
```

```python
sns.jointplot(data = df1, x = 'age', y = 'Mileage')
```

```python
sns.jointplot(data = df1, x = 'age', y = 'amount')
```

```python
sns.jointplot(data = df1, x = 'age', y = 'amount', hue = 'type_of_contract2')
```

```python
# 변수를 일일이 확인하지 않고, 쉽게 전체적으로 확인할 수 있는 함수
sns.pairplot(df1)
```

```python
sns.pairplot(df1[['age', 'Mileage', 'amount', 'product']], hue  = 'product')
```

```python
df1.corr() 
# 0에서 1의 값을 가지고 1로 갈수록 양의 상관관계를 가지고,
# -1로 갈수록 음의 상관고나계를 가진다.
# corr를 확인한 후에 분석하는데 시간을 많이 줄여주게 된다.
```
